```python
import json
import html
import re

from selenium import webdriver
from selenium.webdriver.chrome.options import Options

def get_dynamic_aspx_page_source(url):
    options = Options()
    options.add_argument('--headless')
    driver = webdriver.Chrome(options=options)

    driver.get(url)
    html_source = driver.page_source
    driver.quit()
    return html_source.splitlines()

def find_index_starting_with(strings, prefix):
    for i, s in enumerate(strings):
        if s.startswith(prefix):
            return i
    return -1  # Not found



```

```python

url = "https://medpolicy.ibx.com/ibc/Commercial/Pages/Policy-Bulletin-View.aspx"
html_lines = get_dynamic_aspx_page_source(url)
```

```python
#html_lines = html_source.splitlines()
```

```python
def get_json_with_links(html_lines):
    begin_index = find_index_starting_with(html_lines, 'var WPQ3ListData = ')
    end_index = find_index_starting_with(html_lines, '],"FirstRow"')
    json_lines = html_lines[begin_index+1: end_index]
    json_lines.append("]")
    return '\n'.join(json_lines)

```

```python
json_with_links = get_json_with_links(html_lines)

import json


# Output: {'name': 'Alice', 'age': 30, 'skills': ['Python', 'SQL']}
```

```python
json_with_links
```

```python

all_data =  json.loads(json_with_links)
tuples = []
for data in all_data:
    

    # Step 2: Pull base data
    title = data["MPTitle"].replace('\u200b', '')
    policy_number = data["MPPolicyNumberwithoutversion"]
    main_page = data["MPPolicyPageName"]

    # Step 3: Build tuples
    

    # Main title tuple
    tuples.append((title, main_page))

    # # Step 1: Extract inner HTML content and decode it
    try:
        raw_html = data["MPattachmentdataPub"]
        json_str = re.sub(r'<.*?>', '', raw_html)  # Strip HTML tags
        print(json_str)
        json_str = html.unescape(json_str)        # Convert HTML entities
        attachments = json.loads(json_str)

        # Attachment tuples
        for att in attachments.values():
            letter = att["MPAttachmentLetter"]
            att_title = att["Title"].replace("\\u002f", "/")
            page_id = att["PolicyAttachmentPageName"]

            if "Always Bundled" in att_title:
                display_title = f"Attachment {letter} ({att_title}) to {policy_number}x {title} {policy_number}x"
            else:
                display_title = f"Attachment {letter} ({att_title}) to {policy_number}x {title}"

            tuples.append((display_title, page_id))
    except Exception:
        pass

# Print output
for t in tuples:
    print(t)
```
